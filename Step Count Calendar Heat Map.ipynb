```python
import pandas as pd
import numpy as np; np.random.seed(sum(map(ord, 'calmap')))
import calmap
import matplotlib.pyplot as plt
```

```python
steps = pd.read_csv('./data/StepCount.csv')
steps['Date'] = steps['endDate'].str[0:10]
steps = steps[['value','Date']]
steps = steps.groupby(by='Date')
```

```python
totals = steps.sum()
totals = totals.set_index(pd.DatetimeIndex(pd.to_datetime(totals.index.values)))
totals.head()
```

```python
c = calmap.calendarplot(pd.Series(totals['value'].iloc[0:]),vmin=0,vmax=20000,fig_kws=dict(figsize=(12, 10)))
```

```python
c[0].savefig("StepCount.png")
```
